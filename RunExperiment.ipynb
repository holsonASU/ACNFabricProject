```python
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager
from ipaddress import ip_address, IPv4Address, IPv4Network
import ipaddress
import time
import threading

fablib = fablib_manager()

fablib.show_config();
```

```python
slice_name = 'ProjectSlice'

client_name = 'Client'
graphics_name = 'Graphics'
logic1_name = 'Logic1'
logic2_name = 'Logic2'
logic3_name = 'Logic3'

network1_name = 'net1'
network2_name = 'net2'
network3_name = 'net3'

experimentDirectory = "/home/ubuntu/Experiment/"

clientPython = "client.py"
clientCommandFile = "clientCommands2.txt"

logicPython = "logicserver.py"

graphicPython = "graphicserver.py"

createDirectory = "mkdir {}".format(experimentDirectory)
createDirectory2 = "mkdir {}{}".format(experimentDirectory, 'logs')

slice = fablib.get_slice(slice_name)
```

```python
#set up client

client = slice.get_node(name=client_name)
directory = experimentDirectory + clientCommandFile
client.execute("sudo apt install -y python3")
client.execute("wget https://bootstrap.pypa.io/get-pip.py")
client.execute("python3 ./get-pip.py")
client.execute("python3 -m pip install -r requirements.txt")
client.execute("python3 -m pip install mlsocket")
client.execute(createDirectory)
client.execute(createDirectory2)
client.upload_file(clientCommandFile, directory)
client.upload_file('clientOutput.txt', experimentDirectory+'clientOutput.txt')
client.upload_file(clientPython, experimentDirectory+clientPython)
```

```python
#set up logics
logic1 = slice.get_node(name=logic1_name)
logic1.execute(createDirectory)
logic1.execute(createDirectory2)
logic1.execute("sudo apt install -y python3 pip")
logic1.upload_file(logicPython, experimentDirectory+logicPython)
logic1.upload_file('logicLog.txt', experimentDirectory+'logs/logicLog.txt')
logic1IP = logic1.get_interface(network_name=network1_name).get_ip_addr().compressed

logic2 = slice.get_node(name=logic2_name)
logic2.execute(createDirectory)
logic2.execute(createDirectory2)
logic2.execute("sudo apt install -y python3 pip")
logic2.upload_file(logicPython, experimentDirectory+logicPython)
logic2.upload_file('logicLog.txt', experimentDirectory+'logs/logicLog.txt')
logic2IP = logic2.get_interface(network_name=network2_name).get_ip_addr().compressed

logic3 = slice.get_node(name=logic3_name)
logic3.execute(createDirectory)
logic3.execute(createDirectory2)
logic3.execute("sudo apt install -y python3 pip")
logic3.upload_file(logicPython, experimentDirectory+logicPython)
logic3.upload_file('logicLog.txt', experimentDirectory+'logs/logicLog.txt')
logic3IP = logic3.get_interface(network_name=network3_name).get_ip_addr().compressed
```

```python
#set up graphics
graphics = slice.get_node(name=graphics_name)
graphics.execute(createDirectory)
graphics.execute(createDirectory2)
graphics.execute('mkdir {}/Videos'.format(experimentDirectory))
graphics.execute("sudo apt install -y python3")
graphics.execute("wget https://bootstrap.pypa.io/get-pip.py")
graphics.execute("python3 ./get-pip.py")
graphics.execute("python3 -m pip install -r requirements.txt")
graphics.execute("python3 -m pip install opencv-python-headless")
graphics.execute("python3 -m pip install mlsocket")
#graphics.execute("sudo apt install -y python3-opencv")
graphics.upload_file(graphicPython, experimentDirectory+graphicPython)
graphics.upload_file('graphicsCommandLog.txt', experimentDirectory+'logs/graphicsCommandLog.txt')
graphicsIP = graphics.get_interface(network_name=network1_name).get_ip_addr().compressed
graphics.upload_file('1080.png', experimentDirectory+'1080.png')
graphics.upload_file('720.png', experimentDirectory+'720.png')
```

```python
def startLogic(node, script, directory, quiet, timeout):
    command = "timeout {} python3 {}".format(timeout, directory+script)
    node.execute(command, quiet=quiet)
    node.execute("sudo lsof -i:12344")
    node.execute("sudo lsof -i:12345")

def startGraphics(node, script, directory, logicIP, quiet, timeout, imageFile, experiment, meanGPUTime, stdGPUTime):
    command = "timeout {} python3 {} {} {} {} {}".format(timeout, directory+script, logicIP, directory+imageFile, meanGPUTime, stdGPUTime)
    outputFile = "Logs/graphic{}.txt".format(experiment)
    node.execute(command, quiet=quiet, output_file=outputFile)

def startClient(node, script, commandList, directory, logicIP, graphicIP, quiet, timeout, experiment):
    command = "timeout {} python3 {} {} {} {}".format(timeout, directory+script, directory+commandList, logicIP, graphicIP, "Logs/clientLog{}.txt".format(experiment))
    outputFile = "Logs/client{}.txt".format(experiment)
    node.execute(command, quiet=quiet, output_file=outputFile)
    node.execute("")
```

```python
experiments = [
    [logic1, logic1IP, "1080.png", "8.33", "0.32"],
    [logic1, logic1IP, "1080.png", "16.66", "0.64"],
    [logic1, logic1IP, "720.png", "8.33", "0.32"],
    [logic1, logic1IP, "720.png", "16.66", "0.64"],
    [logic2, logic2IP, "1080.png", "8.33", "0.32"],
    [logic2, logic2IP, "1080.png", "16.66", "0.64"],
    [logic2, logic2IP, "720.png", "8.33", "0.32"],
    [logic2, logic2IP, "720.png", "16.66", "0.64"],
    [logic3, logic3IP, "1080.png", "8.33", "0.32"],
    [logic3, logic3IP, "1080.png", "16.66", "0.64"],
    [logic3, logic3IP, "720.png", "8.33", "0.32"],
    [logic3, logic3IP, "720.png", "16.66", "0.64"]
]
```

```python
if __name__ =="__main__":
    
    experimentNum = 11
    
    experiment = experiments[experimentNum]

    logic = experiment[0]

    logicIP = experiment[1]

    t1 = threading.Thread(target=startLogic, args=(experiment[0], logicPython, 
                                                   experimentDirectory, False,
                                                  90))
    t2 = threading.Thread(target=startGraphics, args=(graphics, graphicPython,
                                                      experimentDirectory,
                                                      logicIP, True,
                                                     90, experiment[2], experimentNum, experiment[3], experiment[4]))
    t3 = threading.Thread(target=startClient, args=(client, clientPython, clientCommandFile,
                                                   experimentDirectory,
                                                   logicIP, graphicsIP, True,
                                                   90, experimentNum))

    t1.start()
    time.sleep(0.5)
    t2.start()
    time.sleep(0.5)
    t3.start()

    t1.join()
    t2.join()
    t3.join()
    
    print("Done Running Experiments")
    
```
